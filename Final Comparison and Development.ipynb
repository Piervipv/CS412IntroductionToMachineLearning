```python
# dataframe management
import pandas as pd             

# numerical computation
import numpy as np



# seaborn can generate several warnings, we ignore them
import warnings 
warnings.filterwarnings("ignore")


from sklearn.model_selection import cross_val_score
import sklearn.metrics as mt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
```

```python
#Load the dataset from Preprocessinc1.ipynb
print("Loading the Dataset from Preprocessing1.ipynb.. ")
dataset = pd.read_csv("./BagOfWordDataSet1.csv")
```

```python
dataset.head()
```

```python
dataset = dataset.drop('Unnamed: 0', 1)
dataset.head()
```

```python
y = dataset['TARGET']
variables = dataset.columns[dataset.columns!='TARGET']
y.head()
```

```python
X = dataset[variables]
X.head()
```

```python
#splitting train and test set in a stratified manner
print("splitting train and test set in a stratified manner..")
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123, stratify = y)
```

#### Now we can perform Oversampling on the training set:

```python
from pandas import *
df_train= concat([X_train, y_train], axis=1, sort=False)
df_train.head()
```

```python
# Class count
count_class_0, count_class_1 = df_train.TARGET.value_counts()
print("negative class instances:")
print(count_class_0)
print("positive class instances:")
print(count_class_1)

print("percentange of positive: " + str(round(count_class_1*100/(count_class_0 + count_class_1),2)))
```

```python
# Divide by class
df_class_0 = df_train[df_train['TARGET'] == 0]
df_class_1 = df_train[df_train['TARGET'] == 1]
```

```python
print("performing Oversampling..")
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
train_over = pd.concat([df_class_0, df_class_1_over], axis=0)

count_class_0, count_class_1 = train_over.TARGET.value_counts()
print("negative class instances:")
print(count_class_0)
print("positive class instances:")
print(count_class_1)

print("percentange of positive: " + str(round(count_class_1*100/(count_class_0 + count_class_1),2)))
```

```python
#split again in X_train, y_train
y_train = train_over['TARGET']
variables = train_over.columns[dataset.columns!='TARGET']
X_train = train_over[variables]
```

```python
def PrintConfusionMatrix(model, true_y, predicted_y, positive=1, negative=-1):
    cm = confusion_matrix(true_y,predicted_y)
    print("\t"+str(model.classes_[0])+"\t"+str(model.classes_[1]))
    print(str(model.classes_[0]) + "\t",cm[0][0],"\t",cm[0][1])
    print(str(model.classes_[1]) + "\t",cm[1][0],"\t",cm[1][1])
```

```python
def F1Score ( precision, recall):
    f1 = 2 * precision * recall / (precision + recall)
    return f1
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
print()
print("Logistic Regression")
ln = LogisticRegression(solver = 'liblinear')
ln.fit(X_train, y_train);
predictions_logistic = ln.predict(X_train)
# check  training accuracy
import sklearn.metrics as mt
acc = mt.accuracy_score(predictions_logistic,y_train)
print( "Training Accuracy: " + str(acc))
```

```python
from sklearn.model_selection import cross_val_score
predictions_logistic = ln.predict(X_test)
accTest = mt.accuracy_score(predictions_logistic,y_test)
print("Test Accuracy on original distribution: " + str(accTest))

scores = cross_val_score(ln, X_train, y_train, cv=10)
#print(scores)
print("performing Cross Validation...")
print("Test Accuracy with CV: " + str(scores.mean()))
print("var: " + str(scores.var()))
```

```python
print("Confusion Matrix - Logistic Regression")
PrintConfusionMatrix(ln, y_test, predictions_logistic)
```

```python
print("Precision %3.2f" % precision_score(y_test,predictions_logistic))
print("Recall    %3.2f" % recall_score(y_test,predictions_logistic))
print("F1    %3.2f" % F1Score( precision_score(y_test,predictions_logistic), recall_score(y_test,predictions_logistic)))
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
print()
print("Ridge:")

    
#Train
clf = RidgeClassifier(alpha=20)
clf.fit(X_train, y_train)
print("alpha: " + str(20))
print()
predictions_ridge = clf.predict(X_train) 
acc = mt.accuracy_score(predictions_ridge,y_train)
print( "Training Accuracy: " + str(acc))
    
#Test on original distribution
predictions_ridge = clf.predict(X_test)
accTest = mt.accuracy_score(predictions_ridge,y_test)
print("Test Accuracy on original distribution: " + str(accTest))

#CV
scores = cross_val_score(clf, X_train, y_train, cv=10)
print("performing Cross Validation...")
print("Test Accuracy with CV: " + str(scores.mean()))
print("var: " + str(scores.var()))


    
print("Precision on test %3.2f" % precision_score(y_test,predictions_ridge))
print("Recall on test  %3.2f" % recall_score(y_test,predictions_ridge))
f1score = F1Score( precision_score(y_test,predictions_ridge), recall_score(y_test,predictions_ridge))
print("F1    %3.2f" % f1score )

print()
```

### Comparison: 
(results can slightly vary for different runs)

#### Accuracy on original Distribution
Logistic Regression: 0.926 <br>
Ridge: 0.93

##### Accuracy with CV
Both Accuracy are pretty high but keep in mind that the original Datased was skewed and before performing Oversampling on the training set we had only 6.20 % of positive instances. Again this means that a classifier that predicts everithing as negative will achieve an Accuracy of 93.8%.
Here the Logistic Regression classifier shows an Accuracy of  0.975% evaluated with 10 Folds CrossValidation.
On the other hand Ridge has an Accuracy of  0.968% evaluated with the same criteria.
Both variances are small and not particularly significative.
Finally remember that these accuracies are computed on the oversampled distribution so they are not so significative.

##### Precision, Recall and F1Score
Here we notice that Ridge has both Precision and Recall greater than the ones of Logistic Regression (and consequently also the f1Score).
Logistic Regression: Precision  0.44; Recall    0.58; F1    0.50.
Ridge:               Precision  0.46; Recall    0.61; F1    0.53.
Again, Precision focuses on the percentage of correctly classified positive examples and Recall focuses on the percentage of positively classified examples with respect to the number of existing ones.

$precision = \frac{TP}{TP+FP}$

$recall = \frac{TP}{TP+FN}$

$ F1 = \frac{2 * Precison * Recall}{Precison + Recall}$

In this contex, where the goal is to correctly identify positive instances (Insincere Questions) these 2 metrics are more significative than a mere Accuracy and this is the reason why will pick the Ridge classifier as our model.
