```python
# dataframe management
import pandas as pd             

# numerical computation
import numpy as np

# visualization library
import seaborn as sns
sns.set(style="white", color_codes=True)
sns.set_context(rc={"font.family":'sans',"font.size":24,"axes.titlesize":24,"axes.labelsize":24})   

# seaborn can generate several warnings, we ignore them
import warnings 
warnings.filterwarnings("ignore")

import sklearn.metrics as mt
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
```

```python
def F1Score ( precision, recall):
    f1 = 2 * precision * recall / (precision + recall)
    return f1
```

```python
print("Loading the test set...")
print()
TestSet = pd.read_csv('./testSet.csv')
TestSet.head() #show the first n instances
```

```python
TestSet = TestSet.drop('Unnamed: 0', 1)
TestSet.head()
```

```python
y = TestSet['TARGET']
variables = TestSet.columns[TestSet.columns!='TARGET']
y.head()
```

```python
X = TestSet[variables]
X.head()
```

##### Let's check the Performance of the really first baseline: the Majority voting classifier.
##### In the train the negative instances are around the 94% of the instances, so a test time this classifier will predict everithing as negative.


```python
string ="Let's check the Performance of the really first baseline: the Majority voting classifier. \n In the train the negative instances are around the 94% of the instances, so a test time this classifier will predict everithing as negative."
print(string)
print()
```

```python
print(" Performance of the really first baseline: the Majority voting classifier:")
print()
```

```python
#let's take the predictions for majority classifiers
predictions_majorityVoting = np.zeros((y.size,), dtype=np.int)
```

```python
acc_MV = mt.accuracy_score(predictions_majorityVoting,y)
print( "Accuracy: " + str(acc_MV))
```

```python
print("Precision %3.2f" % precision_score(y,predictions_majorityVoting))
print("Recall    %3.2f" % recall_score(y,predictions_majorityVoting))
print("F1    %3.2f" % F1Score( precision_score(y,predictions_majorityVoting), recall_score(y,predictions_majorityVoting)))
```

##### Now let's check the performance of a better baseline: NaiveBayes classifier trained over a dataset with the first basic preprocessing and without oversampling. In this way we can check then, how a more sophisticated preprocessing (take a look at Preprocessesing1.py) and Oversampling helped in the task,

```python
#to perform predictions we need to load the testSet with same instances but different preprocessing:
TestSet0 = pd.read_csv('./testSet0.csv')
TestSet0.head() #show the first n instances
```

##### Again: note how this bagOfWord model includes also a lot of irrelevant words

```python
TestSet0 = TestSet0.drop('Unnamed: 0', 1)
y0 = TestSet0['TARGET']
variables0 = TestSet0.columns[TestSet0.columns!='TARGET']
X0 = TestSet0[variables0]
X0.head()
```

```python
print(" Performance of the second baseline:NaiveBayes on the basic preprocess and no OverSampling:")
print()
#load 2nd baseline
from joblib import load
model = load('NaiveBayesBaseline.joblib')
```

```python
predictions_naiveBayes = model.predict(X0)
```

```python
acc_NB = mt.accuracy_score(predictions_naiveBayes,y0)
print( "Accuracy: " + str(acc_NB))
```

```python
print("Precision %3.2f" % precision_score(y0,predictions_naiveBayes))
print("Recall    %3.2f" % recall_score(y0,predictions_naiveBayes))
print("F1    %3.2f" % F1Score( precision_score(y0,predictions_naiveBayes), recall_score(y0,predictions_naiveBayes)))
```
